```python
import pandas as pd
import requests 
import re
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_telephone_operating_companies"
page_response = requests.get(url)
page_info = BeautifulSoup(page_response.text, "html")
```

```python
entire_table = page_info.find_all("table")[1]
print(entire_table)
```

```python
table_titles = entire_table.find_all("th")
print(table_titles)
```

```python
column_titles = [titles.text.strip() for titles in table_titles]
print(column_titles)
```

```python
df = pd.DataFrame(columns=column_titles)
df
```

```python
colomns_data = entire_table.find_all("tr")
colomns_data
```

```python
for row in colomns_data[1:]:
    row_data = row.find_all("td")
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_row_data
df.head()
```

```python
df["Total revenue (US$ billion)"] = df["Total revenue (US$ billion)"].str.replace(r"[\$\[\]]", "", regex=True, flags=re.MULTILINE)
df["Total revenue (US$ billion)"] = pd.to_numeric(df["Total revenue (US$ billion)"])
df.head()
```

```python
df.describe(include="all")
```

```python
df.to_csv("telecommunication_companies.csv", index=False)
```
